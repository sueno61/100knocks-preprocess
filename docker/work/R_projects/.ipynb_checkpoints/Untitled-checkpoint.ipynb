```r
library(readr)
body_data <- read_csv("body_sample.csv")
# View(body_data)
head(body_data)
```

```r
library(ggplot2)
ggplot(body_data, aes(x=height, fill=gender)) +
   geom_histogram() +
   theme_bw(16) +
   ylab("count")
```

```r
ggplot(body_data, aes(x=gender, y=height, fill=gender)) +
  geom_boxplot() +
  theme_bw(16)
```

```r
ggplot(body_data, aes( x=height, y=weight, col=gender) )+   #x軸とy軸を指定
  geom_point(colour="purple", size=3, alpha=0.7)+  #色,サイズ,透明度
  stat_smooth(method="lm", se=T)                   #回帰直線を加える
# seは回帰直線を描く際に上下に幅（95％信頼区間）を持たせる指定
# se=Tで直線に幅を加えて描く（se=Fで直線のみを描く）  
```

```r
ggplot(body_data, aes( x=height, y=weight, col=gender) )+   
  geom_point()+
  theme_bw(16)+
  geom_smooth(method="lm")                   #回帰直線を加える
```
